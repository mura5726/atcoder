```python
N = int(input())
cake = 4
dont = 7
max_kosuu_cake = N // cake
max_kosuu_dont = N // dont
ans = 'No'
for i in range(max_kosuu_cake+1):
    for j in range(max_kosuu_dont+1):
        if cake * i + dont * j == N:
            ans = 'Yes'
print(ans)
```

```python
N, M = map(int, input().split())
A = list(map(int, input().split()))
ans = 0
for i in range(1,N+1):
#     print(i)
    for j in range(N-i+1):
        ij_sum = sum(A[i:i+j])
        if ij_sum != 0:
            if ij_sum % M == 0:
                ans += 1
print(ans)
```

```python
import numpy as np
N, M = map(int, input().split())
A = list(map(int, input().split()))
l = np.arange(1,N+1)
ans = 0
for i in range(0, len(l)):
    for j in range(i, len(l)):
        if sum(A[l[i]-1:l[j]]) % M == 0:
            ans += 1
print(ans)
```

```python
N = int(input())
import math
import itertools
f = []
max_len = int(math.log(abs(N), 2) // 1)
for i in range(max_len+1):
    f.append((-2) ** i)
for m in range(1,max_len+1):
    l = list(itertools.combinations(f,m))
    ans = [0] * len(l)
    for k in range(len(l)):
        if sum(l[k]) == N:
            for j in range(len(l[0])):
                a = [i for i, x in enumerate(f) if x == l[k][j]][0]
                ans[a] = 1
print(''.join(list(map(str,ans[::-1]))))
```
